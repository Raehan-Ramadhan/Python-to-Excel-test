```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
turns = np.arange(1,15)

bank_Cat = [59,68,97,138,293,147,-273,-181,1972,2016,392,1168,1718,2313]
bank_Ice = [572,272,287,230,208,242,32,73,1644,1765,1765,656,378,262]


fig = plt.figure()
plt.plot(turns, bank_Cat)
plt.plot(turns, bank_Ice)

plt.xlabel('Turns')
plt.ylabel('Dollars (in thousands)')
plt.title('Bank Updates by Turns')

plt.legend(['Holy Cat Empire', 'Ice Empire'])

plt.show()
```
