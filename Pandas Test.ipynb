```python
with open('ResourceList.txt') as f:
    lines = f.readlines()
    
Cities = [line.split('~')[0].strip() for line in lines]

Oil = []
for line in lines:
    if (line.find('oil') != -1):
        Oil.append(line[line.find('oil')-4:line.find('oil')-1])
    else:
        Oil.append(np.nan)
        
Stone = []
for line in lines:
    if (line.find('stone') != -1):
        Stone.append(line[line.find('stone')-4:line.find('stone')-1])
    else:
        Stone.append(np.nan)
        
Copper = []
for line in lines:
    if (line.find('copper') != -1):
        Copper.append(line[line.find('copper')-4:line.find('copper')-1])
    else:
        Copper.append(np.nan)
        
Wood = []
for line in lines:
    if (line.find('wood') != -1):
        Wood.append(line[line.find('wood')-4:line.find('wood')-1])
    else:
        Wood.append(np.nan)
        
Tungsten = []
for line in lines:
    if (line.find('tungsten') != -1):
        Tungsten.append(line[line.find('tungsten')-4:line.find('tungsten')-1])
    else:
        Tungsten.append(np.nan)
        
Aluminum = []
for line in lines:
    if (line.find('aluminum') != -1):
        Aluminum.append(line[line.find('aluminum')-4:line.find('aluminum')-1])
    else:
        Aluminum.append(np.nan)
        
Gold = []
for line in lines:
    if (line.find('gold') != -1):
        Gold.append(line[line.find('gold')-4:line.find('gold')-1])
    else:
        Gold.append(np.nan)
        
Iron = []
for line in lines:
    if (line.find('iron') != -1):
        Iron.append(line[line.find('iron')-4:line.find('iron')-1])
    else:
        Iron.append(np.nan)
        
Uranium = []
for line in lines:
    if (line.find('uranium') != -1):
        Uranium.append(line[line.find('uranium')-4:line.find('uranium')-1])
    else:
        Uranium.append(np.nan)
```

```python
pd.DataFrame(list(zip(Iron, Oil, Stone, Aluminum, Gold, Tungsten, Uranium, Copper, Wood)), index=Cities, columns=["Iron", "Oil", "Stone", "Aluminum", "Gold", "Tungsten", "Uranium", "Copper", "Wood"])
```

```python
OilB = []
OilCount = 0
for x in Oil:
    if float(x) > 0.1:
        OilB.append('oil mine')
        OilCount += 1
    else:
        OilB.append(' ')

StoneB = []
StoneCount = 0
for x in Stone:
    if float(x) > 0.1:
        StoneB.append('quarry')
        StoneCount += 1
    else:
        StoneB.append(' ')
        
CopperB = []
CopperCount = 0
for x in Copper:
    if float(x) > 0.1:
        CopperB.append('copper mine')
        CopperCount += 1
    else:
        CopperB.append(' ')
        
WoodB = []
WoodCount = 0
for x in Wood:
    if float(x) > 0.1:
        WoodB.append('logging')
        WoodCount += 1
    else:
        WoodB.append(' ')
        
IronB = []
IronCount = 0
for x in Iron:
    if float(x) > 0.1:
        IronB.append('iron mine')
        IronCount += 1
    else:
        IronB.append(' ')
        
GoldB = []
GoldCount = 0
for x in Gold:
    if float(x) > 0.1:
        GoldB.append('gold mine')
        GoldCount += 1
    else:
        GoldB.append(' ')
        
TungstenB = []
TungstenCount = 0
for x in Tungsten:
    if float(x) > 0.1:
        TungstenB.append('tungsten mine')
        TungstenCount += 1
    else:
        TungstenB.append(' ')
        
AluminumB = []
AluminumCount = 0
for x in Aluminum:
    if float(x) > 0.1:
        AluminumB.append('aluminum mine')
        AluminumCount += 1
    else:
        AluminumB.append(' ')
        
UraniumB = []
UraniumCount = 0
for x in Uranium:
    if float(x) > 0.1:
        UraniumB.append('uranium mine')
        UraniumCount += 1
    else:
        UraniumB.append(' ')
```

```python
mining = pd.DataFrame(list(zip(IronB, OilB, StoneB, AluminumB, GoldB, TungstenB, UraniumB, CopperB, WoodB)), index=Cities, columns=["Iron", "Oil", "Stone", "Aluminum", "Gold", "Tungsten", "Uranium", "Copper", "Wood"])
mining
```

```python
mining.to_excel("output.xlsx", sheet_name='Sheet1')
mining.to_csv("output.csv")
```
